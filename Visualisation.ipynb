```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline

# SETTING STYLES.
#plt.style.use('classic')

import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
plt.show()
```

```python
# SAVING FIGURES TO FILE.
fig.savefig('my_figure.png')
```

```python
# creating panels two for that mater.

plt.figure()

# ( rows, columns, panel number)

plt.subplot(2, 1, 1) 
plt.plot(x, np.sin(x))

# creating the second panel and setting current axis.
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# OBJECT ORIENDTED INTERFACE.
# FIRST CREATE GRID OF PLOTS.
# AX WILL BE AN ARRAY OF TWO AXES OBJECTS.

fig, ax = plt.subplots(2)

# CALL PLOT() METHOD ON THE APPROPRIATE OBJECT.
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

```

```python
# SIMPLE LINE PLOTS.
plt.style.use('seaborn-whitegrid')

# create a matplot figure and an axes.

fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));

# SINGLE FIGURE MULTIPLE VALUES.
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
#ADJUSTING THE PLOT: LINE COLORS AND STYLES.

plt.plot(x, np.sin(x -1), color='blue')
plt.plot(x, np.sin(x -2), color='g')
plt.plot(x, np.sin(x -3), color='0.75')
plt.plot(x, np.sin(x -4), color='#FFDD44')
plt.plot(x, np.sin(x -5), color='chartreuse')
plt.plot(x, np.sin(x -0), color=(1.0, 0.2, 0.3))

```

```python
# ADJUSTING FOR LINE STYLE.

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted');

# OR SHORTENED STYLE.
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':');
```

```python
#ADJUSTING THE PLOT: AXES LIMITS
# USE PLT.XLIM, PLT.YLIM

plt.plot(x, np.cos(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)

# PLT . AXIS ALSO DOES SOME AMAZING STUFF LIKE THE ONE ABOVE AND SO MUCH MORE.
```

```python
# LABELLING PLOTS.

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("Sin(x)")
```

```python
plt.savefig("sine.png")
```

```python
# A PLOT WITH A LEGEND.

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```
